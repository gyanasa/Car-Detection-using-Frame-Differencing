```python
import cv2
import os
  
# Read the video from specified path
cam = cv2.VideoCapture("car_-_2165 (540p).mp4")
  
try:
      
    # creating a folder named data
    if not os.path.exists('frames_new'):
        os.makedirs('frames_new')
  
# if not created then raise error
except OSError:
    print ('Error: Creating directory of data')
  
# frame
currentframe = 0
  
while(True):
      
    # reading from frame
    ret,frame = cam.read()
  
    if ret:
        # if video is still left continue creating images
        # save frame
        name = 'frames_new' + str(currentframe) + '.png'
        print ('Creating...' + name)
  
        # writing the extracted images
        cv2.imwrite(name, frame)
  
        # increasing counter so that it will
        # show how many frames are created
        currentframe += 2 # i.e. at 30 fps, this advances one second
        cam.set(1, currentframe)
    else:
        break
  
# Release all space and windows once done
cam.release()
cv2.destroyAllWindows()
```
